```python
# https://github.com/JustAnotherArchivist/snscrape
import snscrape.modules.twitter as sntwitter
import pandas as pd
import numpy as np
import re
```

```python
twitter_user = sntwitter.TwitterUserScraper("elonmusk")
tweets = []
# DoÄŸruluÄŸu biraz daha arttÄ±rmak iÃ§in tagleri regex ile siliyoruz
regex = r"@\w+ +"


for i, tweet in enumerate(twitter_user.get_items()):
    if i % 100 == 0:
        print("Getting item:", i)
    
    if i == 1000:
        break
        
    tweet.content = re.sub(regex, "", tweet.content)
    tweets.append(tweet)
```

```python
type(tweets[0])
```

```python
index = 1
for tweet in tweets:
    print(index, " ==> ", str(tweet.content))
    index += 1
```

```python
df = pd.DataFrame(tweets)
df.head()
```

```python
content = df["content"]
len(content)
```

```python
word2idx = {}
idx2word = {}
```

```python
i = 0
for cell in content:
    words = cell.split()
    for word in words:
        if word in word2idx:
            continue
        word2idx[word] = i
        idx2word[i] = word
        i += 1
```

```python
word2idx
```

```python
len(word2idx)
```

```python
icerik_sayisi = len(content)
vocab_size = len(word2idx)
```

```python
icerikler = np.zeros((icerik_sayisi, vocab_size))
```

```python
icerikler.shape
```

```python
for i, cell in enumerate(content):
    words = cell.split()
    for word in words:
        wordIndex = word2idx[word]
        icerikler[i][wordIndex] += 1
```

```python
icerikler[7][:10]
```

```python
doc_frequencies = {}
```

```python
# 1 - 5 dk sÃ¼rebilir 
for word in word2idx:
    doc_count = 0
    for icerik in content:
        if word in icerik:
            doc_count += 1
    
    doc_frequencies[word] = doc_count
```

```python
sorted_frequencies = sorted(doc_frequencies, key=doc_frequencies.get, reverse=True)
```

```python
for word in sorted_frequencies[:10]:
    print(word, doc_frequencies[word])
```

```python
for word in sorted_frequencies[-10:]:
    print(word, doc_frequencies[word])
```

```python
# 1 - 5 dk sÃ¼rebilir
icerikler_tf_idf = np.zeros((icerik_sayisi, vocab_size))
icerik_sayisi = len(content)

for i, icerik in enumerate(icerikler):
    if i % 100 == 0:
        print(f"{i} inci iÃ§erik iÅŸleniyor...")
    for j, tf in enumerate(icerik):
        idf = 0
        if tf > 0:
            idf = np.log(icerik_sayisi / (doc_frequencies[idx2word[j]] + 1)) + 1
        tf_idf_score = tf * idf
        icerikler_tf_idf[i][j] = tf_idf_score
```

```python
def cos_sim(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
```

```python
def icerik_2_str(row_index):
    icerik = df.iloc[row_index]
    tarihi = icerik['date']
    tweet_id = icerik['id']
    begeni = icerik['likeCount']
    content = icerik['content']
    #if len(content) > 150:
    #    content = content[:150] + "..."

    return f"Tarihi: {tarihi} \nTweet ID: {tweet_id} \nBeÄŸeni: {begeni} \nÄ°Ã§erik:\n{content}"
```

```python
def most_similar_contents(tweet_id, benzer_icerik_sayisi):
    icerik_indexi = df[df["id"] == tweet_id].index[0]
    print("Aranan Ä°Ã§erik:")
    print(icerik_2_str(icerik_indexi))
    print()
    print("* " * 30)
    print()

    benzerlikler = []
    sorgu_icerik = icerikler_tf_idf[icerik_indexi]
    for vec in icerikler_tf_idf:
        benzerlik = cos_sim(sorgu_icerik, vec)
        benzerlikler.append(benzerlik)
    
    en_benzer_icerik_indexleri = (- np.array(benzerlikler)).argsort()[:benzer_icerik_sayisi + 1]
    print("Benzer Ä°Ã§erikler:\n")
    for i, icerik_idx in enumerate(en_benzer_icerik_indexleri[1:]):
        print(str(i+1) + ") " + icerik_2_str(icerik_idx))
        print()
        print("- " * 20)
        print()
```
